#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #Changing BGR format to RGB
plt.imshow(new_img)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
plt.imshow(cv2.flip(new_img, 0))
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
plt.imshow(cv2.rectangle(new_img, pt1 = (200,700), pt2 = (600, 380), color = (255, 0, 0), thickness = 10))
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
new_img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img2)
```

```python
vertices = np.array([[245, 700], [400, 400], [600, 700]])

pts = vertices.reshape(-1, 3, 2)
```

```python
pts
```

```python
plt.imshow(cv2.polylines(new_img2, [pts], isClosed = True, color = (0, 0, 255), thickness = 20))
```

### BONUS TASK. Can you figure our how to fill in this triangle? It requires a different function that we didn't show in the lecture! See if you can use google search to find it.

```python
plt.imshow(cv2.fillPoly(new_img2, [pts], color = [0, 0, 255]))
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import cv2
import numpy as np

def circle(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:#Checks for left click of the mouse
        cv2.circle(img, (x, y), 100, color = (255, 0, 0), thickness = 10) #Draw the red circle in the selected point
        
img = cv2.imread('../DATA/dog_backpack.jpg') #Importing the dog image
        
cv2.namedWindow(winname = 'dog') #Naming the window

cv2.setMouseCallback('dog', circle) 

while True:
    
    cv2.imshow('dog', img) #Shows the image
    
    if cv2.waitKey(20) & 0xFF == 27: 
        break #Break the loop, once we press the esc key after 20 secs

cv2.destroyAllWindows()
```
